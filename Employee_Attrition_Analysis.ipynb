```python
import pandas as pd

data = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
#it will show the all data inide the slected excle sheet : - 
data.
```

```python
print(data.columns)
```

```python
# how many employees are leaving the company : - 
print(data["Attrition"].value_counts())
```

```python
import matplotlib.pyplot as plt
data["Attrition"].value_counts().plot(kind='bar')

plt.title("employee attrition")
plt.xlabel("Attrition")
plt.ylabel("employee count")

plt.show
```

```python
# Department wise attrition analysis

department_attrition = data.groupby("Department")["Attrition"].value_counts()

print(department_attrition)
```

```python
import matplotlib.pyplot as plt

data.groupby("Department")["Attrition"].value_counts().unstack().plot(kind='bar')

plt.title("Department Wise Attrition ")
plt.xlabel("Department")
plt.ylabel("Employee Count")

plt.show()
```

```python
# are employees are leaving because of the high work preseuur which is extra overtime 

overtime_attrition = data.groupby("OverTime")["Attrition"].value_counts()
print(overtime_attrition)
```

```python
#the plot chart for to check are emp are leaving because of the high work 
import matplotlib.pyplot as plt

data.groupby("OverTime")["Attrition"].value_counts().unstack().plot(kind='bar')

plt.title("OverTime vs Attrition")
plt.xlabel("OverTime")
plt.ylabel("Employee Count")

plt.show()
```

```python
#chech the reason behine the the employee attriction is PercentSalaryHike or not :- 
salaryhike_attrition = data.groupby('PercentSalaryHike')['Attrition'].value_counts()
print(salaryhike_attrition)
```

```python
import matplotlib.pyplot as plt

data.groupby('PercentSalaryHike')['Attrition'].value_counts().unstack().plot(kind = 'bar')


plt.title('employee attrition because of salary hike issue')
plt.xlabel('emp count')
plt.ylabel('percentage of salary hike')

plt.show()

```
